```python
from sympy import Symbol, diff, oo, Interval, is_increasing, is_decreasing
from sympy.solvers import solve, solveset
from sympy.plotting import plot
```

```python
x = Symbol('x', real = True)
y = x**2 - 6*abs(x) + 8
y
```

```python
# 1. Определить корни
roots = solve(y, x)
roots
```

```python
# 4. Построить график
plot(y)
```

```python
interv_y1=Interval(0, +oo)
interv_y2=Interval(-oo, 0)
```

```python
plot(y, (x, 0, 5))
```

```python
plot(y, (x, -5, 0))
```

```python
solveset(y, x, Interval(0, +oo))
```

```python
solveset(y, x, Interval(-oo, 0))
```

```python
F=diff(y, x)
F
```

```python
root_F1 = solveset(F, x, Interval(0, +oo))
root_F1
```

```python
root_F2 = solveset(F, x, Interval(-oo, 0))
root_F2
```

```python
plot(diff(y))
```

```python
intervals_y=[]#интервалы, на которых функция монотонна
intervals_y.append([-oo, int(root_F2.args[0].evalf())])
intervals_y.append([int(root_F2.args[0].evalf()), 0])
intervals_y.append([0, int(root_F1.args[1].evalf())])
intervals_y.append([int(root_F1.args[1].evalf()), +oo])
intervals_y
```

```python
if is_increasing(y, Interval(intervals_y[0][0], intervals_y[0][1])):
    print(f'На интервале {intervals_y[0]} функция возрастает')
else:
    print(f'На интервале {intervals_y[0]} функция убывает')

if is_increasing(y, Interval(intervals_y[1][0], intervals_y[1][1])):
    print(f'На интервале {intervals_y[1]} функция возрастает')
else:
    print(f'На интервале {intervals_y[1]} функция убывает')
if is_increasing(y, Interval(intervals_y[2][0], intervals_y[2][1])):
    print(f'На интервале {intervals_y[2]} функция возрастает')
else:
    print(f'На интервале {intervals_y[2]} функция убывает')
if is_increasing(y, Interval(intervals_y[3][0], intervals_y[3][1])):
    print(f'На интервале {intervals_y[3]} функция возрастает')
else:
    print(f'На интервале {intervals_y[3]} функция убывает')
```

```python
for i,j in enumerate(roots):
    y_sign = y.subs(x, j-0.0001)
    if y_sign > 0 and i == 0:
        print(f'Функция y > 0 на интервале (-oo, {j})')
    if y_sign < 0 and i == 0:
        print(f'Функция y < 0 на интервале (-oo, {j})')
    if y_sign > 0 and i != 0:
        print(f'Функция y > 0 на интервале ({roots[i-1]}, {j})')
    if y_sign < 0 and i != 0:
        print(f'Функция y < 0 на интервале ({roots[i-1]}, {j})') 
    if y_sign > 0 and i == len(roots)-1:
        print(f'Функция y < 0 на интервале ({roots[i]}, +oo)')
    if y_sign < 0 and i == len(roots)-1:
        print(f'Функция y > 0 на интервале ({roots[i]}, +oo)') 
```
