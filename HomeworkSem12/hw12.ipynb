```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd_df = pd.read_csv('company_sales_data.csv')
pd_df.head(12)
```

```python
# 1. Прочитайте все данные о продажах продукта и покажите их с помощью многострочного графика
month_list = pd_df['month_number']
facecream_list = pd_df['facecream']
facewash_list = pd_df['facewash']
toothpaste_list = pd_df['toothpaste']
bathingsoap_list = pd_df['bathingsoap']
shampoo_list = pd_df['shampoo']
moisturizer_list = pd_df['moisturizer']
```

```python
plt.figure(figsize = (12, 8))
plt.plot(month_list, facecream_list, label = 'Крем для лица',
        color = 'r', linestyle = '--', linewidth = 2, marker = 'o', markerfacecolor = 'r')
plt.plot(month_list, facewash_list, label = 'Умывалка',
        color = 'g', linestyle = '-', linewidth = 2, marker = '8', markerfacecolor = 'g')
plt.plot(month_list, toothpaste_list, label = 'Зубная паста',
        color = 'b', linestyle = '-', linewidth = 2, marker = 'x', markerfacecolor = 'b')
plt.plot(month_list, bathingsoap_list, label = 'Мыло',
        color = 'm', linestyle = '-', linewidth = 2, marker = 'h', markerfacecolor = 'm')
plt.plot(month_list, shampoo_list, label = 'Шампунь',
        color = 'y', linestyle = '--', linewidth = 2, marker = 'd', markerfacecolor = 'y')
plt.plot(month_list, moisturizer_list, label = 'Средство для увлажнения',
        color = 'c', linestyle = '--', linewidth = 2, marker = '*', markerfacecolor = 'c')



plt.title('Продажи товаров за год')
plt.xlabel('Номер месяца')
plt.ylabel('Количество проданных едениц товара')
plt.xticks(month_list)
plt.grid(linestyle = '--')
plt.yticks([ 1000, 1500, 2000, 2500,3000, 4000,6000, 9000, 12000, 14000])
plt.legend()
plt.show()
```

```python
# 2. Прочитайте данные о продажах мыла для купания за все месяцы и покажите их с помощью столбчатой диаграммы.
plt.figure(figsize = (10, 8))
plt.bar(month_list, bathingsoap_list, color = 'b', width = 0.5, label = 'Продажи мыла для купания')
plt.xticks(month_list)

plt.title('Информация о продажах мыла для купания')
plt.ylabel('Количество проданных единиц')
plt.xlabel('Номер месяца')
plt.legend()
plt.grid(linestyle = '--')

plt.show()
```

```python
# 3. Рассчитайте общие данные о продажах за год для каждого продукта и покажите их с помощью круговой диаграммы

product_sum = [sum(facecream_list), sum(facewash_list),sum(toothpaste_list), 
          sum(bathingsoap_list), sum(shampoo_list), sum(moisturizer_list)]
name = ['Крем для лица', 'Умывалка', 'Зубная паста', 'Мыло', 'Шампунь', 'Средство для увлажнения']
plt.figure(figsize = (15, 10))
plt.title('Общие данные о продажах за год для каждого продукта')
plt.pie(product_sum, labels = name, autopct ='%1.2f%%')
plt.show()
```
